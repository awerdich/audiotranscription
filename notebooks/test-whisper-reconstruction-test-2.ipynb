```python
## clean pure speakers and speakers with the highest overlap

def get_highest_match(speaker_matches):
    """
    Returns [speaker, match_n] with the highest match_n.
    If there is a tie for highest, returns None.
    """
    if not speaker_matches:
        return None

    # Find max match_n
    max_value = max(speaker_matches, key=lambda x: x[1])[1]

    # Count how many have that max value
    highest = [item for item in speaker_matches if item[1] == max_value]

    if len(highest) == 1:
        return highest[0]
    else:
        return None
        
wordidx_2speaker_clean ={}
for idx in wordidx_2speaker:
    speakers = wordidx_2speaker[idx]
    if len(speakers) == 1:
        wordidx_2speaker_clean[idx] = speakers[0][0]
    
    else:
        if idx!=0 and idx-1 in wordidx_2speaker_clean:
            old_speaker =  wordidx_2speaker_clean[idx-1]

            ## get the len of the old common sequence
            for speaker in wordidx_2speaker[idx-1]:
                if old_speaker == speaker[0]:
                    old_score= speaker[1]
                    
            ## get the len of the current sequence
            updated_speakers=[]
            for speaker in speakers:
                speaker_name = speaker[0]
                speaker_score = speaker[1]
                if speaker_name==old_speaker:
                    speaker_score = speaker_score+old_score
                updated_speakers.append([speaker_name, speaker_score])
            
            
            potential_next= get_highest_match(updated_speakers)
            if potential_next:
                for speaker in speakers:
                    speaker_name = speaker[0]
                    speaker_score = speaker[1]
                    if speaker_name== old_speaker and speaker_score>potential_next[1]:
                            wordidx_2speaker_clean[idx] = speaker_name
                            break

                    
            # if next_:
            #     wordidx_2speaker_clean[idx] = next_[0]
            wordidx_2speaker[idx] = updated_speakers
            
    
            
```

```python
## create new segments for segments that contain more than one speaker from the list of speakers
def split_by_speaker(segment: dict) -> dict:
    """
    Splits a segment dictionary into separate dictionaries per speaker.
    Fixes the 'text' field accordingly.
    Returns a dictionary keyed by speaker name.
    """
    speakers = {}

    for word in segment["words"]:
        speaker = word["speaker"]

        if speaker not in speakers:
            speakers[speaker] = {
                "start": word["start"],
                "end": word["end"],
                "text": "",
                "words": [],
                "speaker": speaker,
            }

        # Append word
        speakers[speaker]["words"].append(word)
        speakers[speaker]["end"] = word["end"]

    # Fix text field for each speaker
    for speaker_data in speakers.values():
        cleaned_words = [
            w["word"].replace("**", "")  # remove markdown artifacts
            for w in speaker_data["words"]
        ]
        speaker_data["text"] = " ".join(cleaned_words)

    return sorted(speakers.values(), key=lambda x: x["start"])


temp_reconstruction_segments=[]
counter_fixed=0
for segment_idx in range(len(reconstruction_segments)):
    speaker_id =[word['speaker'] for word in reconstruction_segments[segment_idx]['words']]
    speaker_id = count_items(speaker_id)
    if len(speaker_id)==1:
        temp_reconstruction_segments.append(reconstruction_segments[segment_idx])
        continue 

    count=0
    for speaker in speaker_id:
        if speaker in SPEAKERS:
            count+=1
    if  count==1:
        temp_reconstruction_segments.append(reconstruction_segments[segment_idx])
    else:
         # print(f'{segment_idx} is fixed')
         ## split segments
         counter_fixed +=1
         new_segments = split_by_speaker(reconstruction_segments[segment_idx])
         print (segment_idx)
         for segment in new_segments:
             temp_reconstruction_segments.append(segment)
print (f'{counter_fixed} Segments were fixed')
```

```python
import difflib
import re

import pandas as pd
import time
def normalize_with_mapping(text):
    original_tokens = re.findall(r"\S+", text)
    normalized_tokens = [re.sub(r"[^\w]", "", tok).lower() for tok in original_tokens]
    return original_tokens, normalized_tokens


def extract_missing_segments(gt_text, pred_text, include_replacements_as_missing=True):

    gt_orig, gt_norm = normalize_with_mapping(gt_text)
    pred_orig, pred_norm = normalize_with_mapping(pred_text)

    matcher = difflib.SequenceMatcher(None, gt_norm, pred_norm)

    missing = []
    substitutions = []

    for tag, i1, i2, j1, j2 in matcher.get_opcodes():

        if tag == "delete":

            gt_text_seg = " ".join(gt_orig[i1:i2])

            missing.append({
                "text": gt_text_seg,
                "gt_start_idx": i1,
                "gt_end_idx": i2 - 1,
                "gt_token_indices": list(range(i1, i2)),
                "pred_alignment_idx": j1,  # 👈 alignment position in pred
                "op": "delete",
            })

        elif tag == "replace":

            gt_text_seg = " ".join(gt_orig[i1:i2])
            pred_text_seg = " ".join(pred_orig[j1:j2])

            substitutions.append({
                "gt_text": gt_text_seg,
                "pred_text": pred_text_seg,
                "gt_start_idx": i1,
                "gt_end_idx": i2 - 1,
                "pred_start_idx": j1,
                "pred_end_idx": j2 - 1,
                "op": "replace",
            })

            if include_replacements_as_missing and gt_text_seg.strip():
                missing.append({
                    "text": gt_text_seg,
                    "gt_start_idx": i1,
                    "gt_end_idx": i2 - 1,
                    "gt_token_indices": list(range(i1, i2)),
                    "pred_alignment_idx": j1,  # 👈 alignment anchor
                    "op": "replace",
                })

    return {"missing": missing, "substitutions": substitutions}


def find_text_pred(idx, speaker_metadata, missing_segment):
    selected_block= None
    
    if idx == speaker_metadata[-1]['end_token']:
        selected_block = speaker_metadata[-1]
    else:
        for block in speaker_metadata:
            if idx >=block['start_token'] and idx<block['end_token']:
                selected_block = block
                break
    if selected_block:       
        text= selected_block['text']
        index = idx - selected_block['start_token']  
        text = text.split()
        if index ==0:
            missing_segment = f'[{missing_segment} *** OAPS]' ## OR AFTER PREVIOUS SPEECH
        else:
            missing_segment = f'[{missing_segment}]'

        placeholder= '[*******]'

        edited_transcript = text.copy()
        edited_transcript.insert(index, f'{missing_segment}')
        text.insert(index, placeholder)
        start_time = time.strftime('%H:%M:%S', time.gmtime( selected_block['start']))
        end_time = time.strftime('%H:%M:%S', time.gmtime(selected_block['end']))
        dict_ = {'Transcript Text': ' '.join(text), 'Ref-Start Time': start_time, 'Ref-End Time': end_time, 'Edited Transcript': ' '.join(edited_transcript),'block_idx': selected_block['block_idx']}
        return dict_
    return {'Transcript Text': np.nan , 'Ref-Start Time': np.nan, 'Ref-End Time': np.nan, 'Edited Transcript': np.nan, 'block_idx': np.nan}
```

```python
from docx import Document
from docx.shared import Inches, Pt
from docx.oxml.ns import qn
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

def df_to_docx_table(df, filepath, title=None, col_widths=None, fontsize=10):
    """
    Write a pandas DataFrame to a .docx file as a table.

    Args:
        df (pd.DataFrame): DataFrame to write.
        filepath (str): Path to save the .docx file (e.g. "out.docx").
        title (str or None): Optional title to add above the table.
        col_widths (list or None): Optional list of widths in inches for each column.
                                   If provided length must match number of columns.
                                   Example: [1.5, 2.0, 1.0]
        fontsize (int): Font size for table text (points).

    Returns:
        str: filepath saved.
    """
    # lazy import pandas type-check (optional)
    import pandas as pd
    if not hasattr(df, "columns"):
        raise ValueError("df must be a pandas DataFrame")

    # create document
    doc = Document()

    # optional title
    if title:
        p = doc.add_paragraph()
        run = p.add_run(title)
        run.bold = True
        run.font.size = Pt(fontsize + 2)
        p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
        doc.add_paragraph()  # blank line

    # create table (header row + data rows)
    rows = len(df) + 1
    cols = len(df.columns)
    table = doc.add_table(rows=rows, cols=cols)
    table.style = 'Table Grid'  # simple grid style; change if you have other styles

    # header row
    for ci, col_name in enumerate(df.columns):
        cell = table.rows[0].cells[ci]
        cell.text = str(col_name)
        # bold header text and set font size
        for paragraph in cell.paragraphs:
            for run in paragraph.runs:
                run.font.bold = True
                run.font.size = Pt(fontsize)

    # fill data rows
    for r, (_, row) in enumerate(df.iterrows(), start=1):
        for c, col in enumerate(df.columns):
            value = row[col]
            # replace NaN/None with empty string
            if pd.isna(value):
                text = ""
            else:
                text = str(value)
            cell = table.rows[r].cells[c]
            cell.text = text
            # set font size for content
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.size = Pt(fontsize)

    # optional: set column widths (in inches)
    if col_widths is not None:
        if len(col_widths) != cols:
            raise ValueError("col_widths length must match number of DataFrame columns")
        for ci, width_in in enumerate(col_widths):
            for cell in table.columns[ci].cells:
                cell.width = Inches(width_in)
                # also set tcPr w to keep word from auto-wrapping too aggressively
                # (python-docx does not guarantee exact widths in many viewers, but helps)
    # save
    doc.save(filepath)
    return filepath
```

```python
import re
from textblob import TextBlob

## grammar corrections

slang_map = {
        "wanna": "want to",
        "gonna": "going to",
        "gotta": "got to",
        "lemme": "let me",
        "kinda": "kind of",
        "sorta": "sort of",
        "outta": "out of",
        "ain't": "is not",
        "ya": "you",
        "cuz": "because",
        "cause": "because",
        "dunno": "do not know"
    }

def normalize_slang(text, slang_map):
    # Replace whole words only (case insensitive)
    for slang, formal in slang_map.items():
        pattern = r'\b' + re.escape(slang) + r'\b'
        text = re.sub(pattern, formal, text, flags=re.IGNORECASE)
    return text

def correct_spelling(text):
    blob = TextBlob(text)
    corrected_text = str(blob.correct())
    
    original_words = text.split()
    corrected_words = corrected_text.split()
    for orig, corr in zip(original_words, corrected_words):
        if orig != corr:
            print(f"{orig}  →  {corr}")
    return corrected_text

def capitalize_pronoun_i(text):
    """
    Capitalizes standalone 'i' pronoun.
    Does NOT affect words like 'in', 'is', etc.
    """
    return re.sub(r'\bi\b', 'I', text)

def capitalize_sentences(text):
    """
    Capitalizes the first letter after ., ?, or !
    and also capitalizes the first letter of the text.
    """

    # Capitalize first letter of entire text
    text = text.strip()
    if text:
        text = text[0].upper() + text[1:]

    # Capitalize letter after sentence-ending punctuation
    text = re.sub(
        r'([.!?]\s+)([a-z])',
        lambda m: m.group(1) + m.group(2).upper(),
        text
    )

    return text
```

```python
def correct_text(original_text):
    
    ## remove slangs
    text = normalize_slang(original_text, slang_map)

    ### capitalize nouns (I)
    text = capitalize_pronoun_i(text)

    ### capitalize letter at the beginning of sentences
    text = capitalize_sentences(text)
    
    ### correct spelling 
    # text = correct_spelling(text)
    
    return text

```

```python
import re
from docx import Document
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.oxml import OxmlElement
from docx.oxml.ns import qn

def _add_colored_runs(para, text):
    # Split by **...** or *...*
    parts = re.split(r"(\*\*.*?\*\*|\*.*?\*)", text)
    for part in parts:
        if part.startswith("**") and part.endswith("**"):
            run = para.add_run(part[2:-2])
            run.font.color.rgb = RGBColor(255, 0, 0)   # red
        elif part.startswith("*") and part.endswith("*"):
            run = para.add_run(part[1:-1])
            run.font.color.rgb = RGBColor(0, 128, 0)   # green (your comment says ORANGE)
        else:
            para.add_run(part)

def create_word_document_transcriptColored_noTimeStamp(docx_path, output_fileName):
    doc = Document()

    # ---------- ADD PAGE NUMBER FOOTER ----------
    section = doc.sections[0]
    footer = section.footer
    footer_para = footer.paragraphs[0]
    footer_para.alignment = WD_ALIGN_PARAGRAPH.CENTER

    run = footer_para.add_run()
    fldChar_begin = OxmlElement('w:fldChar')
    fldChar_begin.set(qn('w:fldCharType'), 'begin')

    instrText = OxmlElement('w:instrText')
    instrText.text = 'PAGE'

    fldChar_end = OxmlElement('w:fldChar')
    fldChar_end.set(qn('w:fldCharType'), 'end')

    run._r.append(fldChar_begin)
    run._r.append(instrText)
    run._r.append(fldChar_end)
    # -------------------------------------------

    # Timestamp is OPTIONAL now
    pattern = re.compile(r"^(?:\[(.*?)\]\s*)?([^:]+):\s*(.*)$")

    with open(output_fileName, "r", encoding="utf-8") as file:
        for line in file:
            stripped_line = line.strip()
            if not stripped_line:
                continue

            para = doc.add_paragraph()
            para.alignment = WD_ALIGN_PARAGRAPH.JUSTIFY

            m = pattern.match(stripped_line)

            if m:
                timestamp, speaker, text = m.groups()

                # Add timestamp if present
                if timestamp:
                    para.add_run(f"[{timestamp}] ")

                # Speaker in bold
                run_speaker = para.add_run(speaker.strip() + ": ")
                run_speaker.bold = True

                # Color-marked text
                _add_colored_runs(para, text)

            else:
                # No speaker/timestamp format, but STILL color ** ** / * *
                _add_colored_runs(para, stripped_line)

    doc.save(docx_path)
```

```python
import re


def remove_repeated_words(text, threshold=3):
    # word repeated with optional comma/space between
    pattern = re.compile(r'\b(\w+)(?:[,\s]+\1\b){' + str(threshold) + r',}', re.IGNORECASE)

    prev = None
    while prev != text:
        prev = text
        text = pattern.sub(r'\1', text)

    return text




text = """uh uh uh uh uh uh uh uh uh uh And we are doing our best"""
remove_repeated_words(x, threshold=4)
```
